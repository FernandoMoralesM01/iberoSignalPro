El objeto vicon tiene como propiedades distintos dataframes, los cuales son extraidos por los archivos de Vicon.

actualmente el objeto de vicon solo tiene una función que el usuario puede acceder llamada: leeViconCSV.

```python
v = Vicon()
v.leeViconCSV("002.csv")
```

```python
fs = 2000
t = np.arange(0, len(v.devices["IM EMG1"]), 1)/fs
```

```python
sos = sgnl.butter(2, [10*2/fs, 500*2/fs], "bandpass", output = "sos")
emgfilt = sgnl.sosfiltfilt(sos, v.devices["IM EMG1"])
```

```python
emg = ib.normaliza(emgfilt, np.mean(emgfilt), np.std(emgfilt))
env = ib.obtenerEnvolvente(emg)
```

```python
plt.figure(figsize=(20, 5))
plt.plot(t, emg)
plt.plot(t, env)
plt.grid()
```
